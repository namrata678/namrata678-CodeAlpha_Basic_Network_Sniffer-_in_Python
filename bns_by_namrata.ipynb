**TASK 1 BASIC NETWORK SNIFFER**

Build a network sniffer in Python that captures and analyzes network traffic. This project will help you understand how data flows on a network and how network packets are structured.

```python
# Install scapy library
!pip install scapy

# Import necessary modules from scapy library
import scapy.all as scapy

# Function to sniff network traffic on a specified interface
def sniff(interface):
    # Use scapy's sniff function to capture packets
    scapy.sniff(iface=interface, store=False, prn=process_packet)

# Function to process each captured packet
def process_packet(packet):
    # Check if the packet has an IP layer
    if packet.haslayer(scapy.IP):
        # Extract source and destination IP addresses and the IP protocol
        ip_src = packet[scapy.IP].src
        ip_dst = packet[scapy.IP].dst
        protocol = packet[scapy.IP].proto

        # Print information about the new packet
        #Coded by Namrata Ingle
        print(f"\n[+] New Packet: {ip_src} to {ip_dst} (Protocol: {protocol})")

        # Check if the packet has a TCP layer
        if packet.haslayer(scapy.TCP):
            # Extract source and destination TCP port information
            src_port = packet[scapy.TCP].sport
            dst_port = packet[scapy.TCP].dport
            print(f"    TCP - Source Port: {src_port}, Destination Port: {dst_port}")

        # Check if the packet has a UDP layer
        elif packet.haslayer(scapy.UDP):
            # Extract source and destination UDP port information
            src_port = packet[scapy.UDP].sport
            dst_port = packet[scapy.UDP].dport
            print(f"    UDP - Source Port: {src_port}, Destination Port: {dst_port}")

        # Check if the packet has an ICMP layer
        elif packet.haslayer(scapy.ICMP):
            print("    ICMP Packet")

        # If none of the above conditions are met, print a generic message
        else:
            print("    Other IP Packet")

# Call the sniff function with the specified network interface
sniff("eth0")
```
